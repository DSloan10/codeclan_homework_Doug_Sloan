5. Summarise the mean number of calories in each beverage_category. Which 3 category of drinks have the most calories? Which 3 drinks have the least? Write a small summary of your findings.



```python
drinks_avcal_by_cat = drinks.groupby("Beverage_category")["Calories"].mean().reset_index(name = "mean_calories").sort_values("mean_calories")
```

```python
drinks_avcal_by_cat
```

The Beverage categories with the most amount of calories are "Smoothies", "Frappuccino® Blended Coffee" and "Signature Espresso Drinks" in that order. The Beverage categories with the least amount of calories are "Coffee", "Shaken Iced Beverages" and "Classic Espresso Drinks" in that order. It seems reasonable to assume that as drinks become more complex, more elements are added such as syrups, milk and fruit which lead to a larger calorie count. 

6. Harder - Which Beverage_prep type contains more than the average calories of all drinks? (Hint: to answer this, you’ll have to first figure out what the average calories across all drinks are, and then use that as a filter for the grouped Beverage_prep data).



```python
mean_all_drinks = drinks["Calories"].mean()
```

```python
drinks_avcal_by_prep = drinks.groupby("Beverage_prep")["Calories"].mean().reset_index(name = "mean_calories").sort_values("mean_calories")
```

```python
drinks_avcal_by_prep.loc[drinks_avcal_by_prep["mean_calories"] > mean_all_drinks]
```
